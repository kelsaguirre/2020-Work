```python
#import packages
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#define the style like in lecture
plt.style.use('classic')
#I picked Austin, TX as my city
#read in the listings file for Austin
listings =  pd.read_csv('Documents/Data_Science_I/Week_10/Data_Challenge_5/listings.csv')
```

```python
#I picked Austin, TX as my city
#read in the reviews file for Austin
reviews =  pd.read_csv('Documents/Data_Science_I/Week_10/Data_Challenge_5/reviews.csv')
```

## Figure 1

```python
#define the x and y axis
lat = listings['latitude']
lon = listings['longitude']
#define the variable I will use for color
price = listings['price']
#set the variables for the graph
plt.scatter(lon, lat,
            c = price,
            alpha = 0.3)

#label the axis
plt.xlabel('Longitude')
plt.ylabel('Latitude')
#label the colorbar
plt.colorbar(label = 'Price ($)')
#limit the colorbar 
plt.clim(0,11000)
#setting x and y limits to spread out the points
plt.ylim(30.50,30.1)
plt.xlim(-98.057,-97.57)

plt.title('Austin, TX AirBnBs: Location on X and Y and Price via Color')
```

Figure 1: Here I've used the longitude and latitude to plot the geographical location of each Air BnB in Austin, TX. I then added color to see if the lake properties on the outskirts of Austin cost more per night than other AirBnBs. In central Austin it's difficult to interpret the price data because there are too many overlapping data points. I adjusted the intercept limits to spread this out a bit and after that it is clear with this plot is how the nicest part of austin, the Westlake area (southwest) is clearly the most expensive area of Austin.

## Figure 2

```python
#set the two variables for the x and y axis
calculated_host_listings_count = listings['calculated_host_listings_count']
price = listings['price']
#define the variables in the bar graph
plt.bar(calculated_host_listings_count, price,
        #change the color!
        color = 'blue')
#set a limit to make a more compelling plot
plt.xlim(0,150)
#label the axis
plt.xlabel('Number of Listings Host Has')
plt.ylabel('Price ($)')
#label the title
plt.title('Austin, TX AirBnBs: Number of Host Listings vs Price of Listing')
```

Figure 2: Here I plotted the number of listings an individual host has, versus the price of the listing. Before plotting, I was thinking that the more listings an individual host has, the more familiar the host is with the entire AirBnB process and therefore they're more likely to charge more for their listing. I expected a stronger correlation with these two variables, but there still seems to be a positive correlation that oddly drops around 20-40 listings.

## Figure 3

```python
#define the violin plot characteristics and dataset
sns.violinplot(x="room_type",
            y="reviews_per_month",
            data=listings)
#label the title
plt.title('Austin, TX AirBnBs: Room Type vs Number of Reviews per Month')
#label each of the axis
plt.xlabel('Room Type')
plt.ylabel('Reviews Per Month')
```

Figure 3: In this figure I plotted Room Type on the x axis and number of reviews per month on the y axis. I used a violin plot instead of a boxplot ot represent density of each in additio to the other two variables. It seems the majority of the four different room types receive between 0-5 reviews per month. This plot also makes it clear that the Entire home or apartment category gets the most reviews per month versus the other three, with private rooms coming in second. This may be due to both of those being the most popular categories. Another variable or plot would be needed to see if that's indeed the case.

## Figure 5

```python
#merge the two CSV files, keeping all from reviews CSV so that I can look at dates for reviews
#and have the correct dimensions for the plot
merged = pd.merge(listings, reviews, left_on = 'id', right_on = "listing_id", how = "right")

#define the x and y variables
num_of_revs = merged['reviews_per_month']
review_dates = merged['date']
#drop the nas to clean it up
review_dates = review_dates.dropna()


#define the plot
plt.plot(num_of_revs, review_dates)

plt.xlabel('Number of Reviews Per Month')
plt.ylabel('Date of Reviews')
plt.title('Austin, TX AirBnBs: The Number of Reviews Per Month vs Dates for Reviews')
```

Figure 5: For this figure I merged the two datasets and then took the dates for each review and compared it to the number of reviews per month for each listing. You can see that there isn't much to view in this figure because there is too much information. I need to limit the information loaded in order to see a trend here.
